#include <bits/stdc++.h>
using namespace std;
 
void Palind_fun(string text)
{
    int N = text.length();
    if (N == 0)
        return;
    N = 2 * N + 1;
    int L[N];
    L[0] = 0;
    L[1] = 1;
    int C = 1;
    int R = 2;
    int i = 0;
    int var_inverso;
    int var_aum = -1;
    int distinto = -1;
    int LPS_tam = 0;
    int LPS_centro = 0;
    int inicio = -1;
    int fin = -1;

    for (i = 2; i < N; i++) {
        var_inverso = 2 * C - i;
        var_aum = 0;
        distinto = R - i;

        if (distinto >= 0) {

            if (L[var_inverso] < distinto)
                L[i] = L[var_inverso];
              
            else if (L[var_inverso] == distinto && R == N - 1)
                L[i] = L[var_inverso];
 
            else if (L[var_inverso] == distinto && R < N - 1) {
                L[i] = L[var_inverso];

                var_aum = 1;
            }

            else if (L[var_inverso] > distinto) {
                L[i] = distinto;
                var_aum = 1;
            }
        }
        else {
            L[i] = 0;
            var_aum = 1;
        }
 
        if (var_aum == 1) {
 
            while (((i + L[i]) < N && (i - L[i]) > 0)
                   && (((i + L[i] + 1) % 2 == 0)
                       || (text[(i + L[i] + 1) / 2]
                           == text[(i - L[i] - 1) / 2]))) {
                L[i]++;
            }
        }
 
        if (L[i] > LPS_tam) {
            LPS_tam = L[i];
            LPS_centro = i;
        }

        if (i + L[i] > R) {
            C = i;
            R = i + L[i];
        }

    }
 
    inicio = (LPS_centro - LPS_tam) / 2;
    fin = inicio + LPS_tam - 1;
    cout << "El LPS de la cadena es:  " << text << " : ";
    for (i = inicio; i <= fin; i++)
        cout << text[i];
    cout << endl;
}
